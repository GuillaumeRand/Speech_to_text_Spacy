```python

with open('asset/SpeechDestination.csv', 'r') as f:
    sentences = f.read()

# remove first line
sentences = sentences.splitlines()[1:]


for i in range(0 ,100):
    # split lines on ,
    line = sentences[i].split(',')
    # get the first element
    print(line[0])
    with open('asset/sentences.txt', 'a') as f:
        f.write(line[0])
        f.write('\n')
```
