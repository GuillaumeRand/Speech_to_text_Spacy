```python
import pandas as pd
import random

csvLocation = pd.read_csv('../dijkstra/clearData.csv', sep=',', on_bad_lines='skip')
csvSpeech = pd.read_csv('../asset/SpeechDestination.csv', sep=',', on_bad_lines='skip')
```

```python
# get all depart from csv
def get_all_locations(df):
    locations = []
    for index, row in df.iterrows():
        if row['depart'] not in locations:
            locations.append(row['depart'])
        if row['destination'] not in locations:
            locations.append(row['destination'])
    return locations

allLocations = get_all_locations(csvLocation)
```

```python
for i in range(0, 1000):
    rowNum = random.randint(0, len(csvSpeech))
    # select the row of csvSpeech
    rowSpeech = csvSpeech.sample().iloc[0]
    depart, arrive = random.sample(allLocations, 2)
    if rowSpeech['ERROR'] == 'true':
        continue
    print(rowSpeech['DEPART'], depart, rowSpeech['DESTINATION'], arrive)
    text = rowSpeech['TEXT'].replace(rowSpeech['DEPART'], depart).replace(rowSpeech['DESTINATION'], arrive)
    new_row = pd.DataFrame({'TEXT': [text], 'DEPART': [depart], 'DESTINATION': [arrive], 'ERROR': 'false'})
    new_row.to_csv('../asset/SpeechDestination.csv',mode='a',header=False, index=False)

```
